```python
import pandas as pd

# Load the two CSV files into data frames
big_df = pd.read_csv('C:/Users/Dell/Documents/Business Analytics Year 4/Applied Predictive Analytics DATA4001 2022-23/NBA_Player_Stats_2.csv')
small_df = pd.read_csv('C:/Users/Dell/Documents/Business Analytics Year 4/Applied Predictive Analytics DATA4001 2022-23/NBA_Salary_stats_2.csv')

big_df['Player'] = big_df['Player'].str.replace('*', '')
# Make a copy of the big data frame to store the updated data
big_df_updated = big_df.copy()

# Iterate over the rows in the small data frame
for _, small_row in small_df.iterrows():
  # Get the player name and season from the small data frame
  player = small_row['Player']
  season = small_row['Season']

  # Find the rows in the big data frame that match the player and season
  matching_rows = big_df[(big_df['Player'].str.lower() == player.lower()) & (big_df['Season'] == season)]

  # If a match was found, update the big data frame with the salary and inflation-adjusted salary from the small data frame
  if not matching_rows.empty:
    big_df_updated.loc[matching_rows.index, 'Player salary'] = small_row['Player salary']
    big_df_updated.loc[matching_rows.index, 'Salary Adjusted for inflation'] = small_row['Salary Adjusted for inflation']


big_df
```

```python
import pandas as pd

# Load the franchises CSV file into a data frame
franchises_df = pd.read_csv('C:/Users/Dell/Documents/Business Analytics Year 4/Applied Predictive Analytics DATA4001 2022-23/NBA_team_caps2.csv')

# Iterate over the rows in the franchises data frame
for _, franchises_row in franchises_df.iterrows():
  # Get the team name and season from the franchises data frame
  tm = franchises_row['Tm']
  season = franchises_row['Season']

  # Find the rows in the big data frame that match the team and season
  matching_rows = big_df_updated[(big_df_updated['Tm'].str.lower() == tm.lower()) & (big_df_updated['Season'] == season)]

  # If a match was found, update the big data frame with the salary cap and inflation-adjusted salary cap from the franchises data frame
  if not matching_rows.empty:
    big_df_updated.loc[matching_rows.index, 'Salary Cap'] = franchises_row['Salary Cap']
    big_df_updated.loc[matching_rows.index, 'Adjusted for inflation'] = franchises_row['Adjusted for inflation']


```

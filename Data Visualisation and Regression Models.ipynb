```python
# Group data by 'Pos' and 'Season' and calculate the average 'Player salary'
avg_salary_by_pos_season = df.groupby(['Pos', 'Season'])['Player salary'].mean().reset_index()
# Format season string to show only last two digits of starting year
avg_salary_by_pos_season['Season'] = avg_salary_by_pos_season['Season'].apply(lambda x: str(int(x.split('-')[0][-2:])))

# Create a time series graph showing the average 'Player salary' of each 'Pos' across the seasons 'Season'
fig, ax = plt.subplots(figsize=(12, 8))
for pos in avg_salary_by_pos_season['Pos'].unique():
    pos_data = avg_salary_by_pos_season[avg_salary_by_pos_season['Pos'] == pos]
    ax.plot(pos_data['Season'], pos_data['Player salary'], label=pos)
ax.set_xlabel('Season')
ax.set_ylabel('Average Player Salary')
ax.set_title('Average Player Salary by Position and Season')
ax.legend()
ax.yaxis.set_major_formatter('${x:,.0f}')
plt.show()
```

```python
def top_5_highest_paid_players_bar_chart(df):
    # Filter data for season 2021-22
    season_2122_df = df[df['Season'] == '2021-22']
    
    # Sort by player salary in descending order and select top 5
    top_5_df = season_2122_df.sort_values(by='Player salary', ascending=False).head(5)
    
    # Create bar chart
    fig, ax = plt.subplots(figsize=(12, 8))
    bars = ax.bar(top_5_df['Player'], top_5_df['Player salary'])
    
    # Set different color for each bar
    colors = ['#5DA5DA', '#FAA43A', '#60BD68', '#F17CB0', '#B2912F']
    for i in range(len(bars)):
        bars[i].set_color(colors[i])
    
    # Set title and axis labels
    ax.set_title('Top 5 Highest Paid Players in the 2021-22 Season')
    ax.set_xlabel('Player')
    ax.set_ylabel('Player Salary (tens of millions of USD)')
    
    # Rotate x-axis labels for better readability
    plt.xticks(rotation=45, ha='right')
    
    # Add player positions to chart
    for i, bar in enumerate(bars):
        ax.text(bar.get_x() + bar.get_width()/2., 1.05*bar.get_height(), top_5_df.iloc[i]['Pos'], ha='center', va='bottom', fontsize=10)
    
    # Show plot
    plt.show()

top_5_highest_paid_players_bar_chart(df)
```

```python
def top_5_highest_paid_players_bar_chart(df):
    positions = ['PG', 'SG', 'PF', 'SF', 'C']
    colors = ['#5DA5DA', '#FAA43A', '#60BD68', '#F17CB0', '#B2912F']
    
    for pos, color in zip(positions, colors):
        # Filter data for the position and season 2021-22
        pos_season_df = df[(df['Pos'] == pos) & (df['Season'] == '2021-22')]
        
        # Sort by player salary in descending order and select top 5
        top_5_df = pos_season_df.sort_values(by='Player salary', ascending=False).head(5)
        
        # Create bar chart
        fig, ax = plt.subplots(figsize=(12, 8))
        bars = ax.bar(top_5_df['Player'], top_5_df['Player salary'], color=color)
        
        # Set title and axis labels
        ax.set_title(f'Top 5 Highest Paid {pos}s in the 2021-22 Season')
        ax.set_xlabel('Player')
        ax.set_ylabel('Player Salary (tens of millions of USD)')
        
        # Rotate x-axis labels for better readability
        plt.xticks(rotation=45, ha='right')
        
        # Show plot
        plt.show()
top_5_highest_paid_players_bar_chart(df)
```

```python
# Define number of folds for K-fold cross-validation
num_folds = 5

# Define features and target variable
features = ['PTS', 'TRB', 'AST']
target = '%_of_cap'

# Define K-fold cross-validation object
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

# Define lists to store RMSE and R-squared values for each fold
rmse_values = []
r2_values = []

# Split data into training and test sets for each fold
for train_index, test_index in kf.split(newdf):
    X_train, X_test = newdf.iloc[train_index][features], newdf.iloc[test_index][features]
    y_train, y_test = newdf.iloc[train_index][target], newdf.iloc[test_index][target]
    
    # Fit linear regression model to training data
    lin_reg = LinearRegression()
    lin_reg.fit(X_train, y_train)

    # Make predictions on test data
    y_pred = lin_reg.predict(X_test)

    # Evaluate model performance for this fold
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)
    rmse_values.append(rmse)
    r2_values.append(r2)

# Compute mean RMSE and R-squared values across all folds
mean_rmse = np.mean(rmse_values)
mean_r2 = np.mean(r2_values)

print(f'Mean RMSE: {mean_rmse:.4f}')
print(f'Mean R-squared: {mean_r2:.4f}')
```
